```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
sns.set(style="white", color_codes=True)
from sklearn.datasets import load_iris
```

```python
# Convert sklearn dataset to Pandas DataFrame

data = load_iris()
iris_df = pd.DataFrame(data = np.c_[data['data'], [data['target_names'][y] for y in data['target']]],
                     columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'] + ['Species'])
iris_df.head()
```

```python
iris_df["Species"].value_counts()
```

```python
iris_df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] = iris_df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].astype(float)
```

# Basic Plots

```python
# SepalLengthCm  VS  SepalWidthCm

iris_df.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# PetalLengthCm  VS  PetalWidthCm

iris_df.plot(kind="scatter", x="PetalLengthCm", y="PetalWidthCm")
```

```python
# Seaborn's JointPlot

sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris_df, size=7)
```

```python
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=iris_df, size=7)
```

```python
# seaborn's FacetGrid to color the scatterplot by 'Species'

sns.FacetGrid(iris_df, hue="Species", size=7) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
sns.FacetGrid(iris_df, hue="Species", size=7) \
   .map(plt.scatter, "PetalLengthCm", "PetalWidthCm") \
   .add_legend()
```

```python
# Visualize inidividual feature corresponding to each Species
sns.boxplot(x="Species", y="PetalWidthCm", data=iris_df)
```

```python
# Overlapping multiple plots
ax = sns.boxplot(x="Species", y="PetalWidthCm", data=iris_df)
ax = sns.stripplot(x="Species", y="PetalWidthCm", data=iris_df, jitter=True, edgecolor="gray")
```

```python
# Denser regions of the data are fatter, and sparser thiner in a seaborn Violin plot
sns.violinplot(x="Species", y="PetalWidthCm", data=iris_df, size=7)
```

```python
# Seaborn's KDE plot visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(iris_df, hue="Species", size=7) \
   .map(sns.kdeplot, "PetalWidthCm") \
   .add_legend()
```

```python
# Seaborn's pairplot shows the bivariate relation between each pair of features
sns.pairplot(iris_df, hue="Species", size=3)
```

```python
sns.pairplot(iris_df, hue="Species", size=3, diag_kind="kde")
```

```python
# Pandas dataframe boxplot for each feature 
iris_df.boxplot(by="Species", figsize=(12, 6))
```

# Sophesticated Plots

```python
# Andrews Curves -> https://en.wikipedia.org/wiki/Andrews_plot
from pandas.tools.plotting import andrews_curves
andrews_curves(iris_df, "Species")
```

```python
# parallel_coordinates is another multivariate pandas' visualization technique which plots each feature on a 
# separate column & then draws lines connecting the features for each data sample
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(iris_df, "Species")
```

```python
# Yet another pandas' multivariate visualization technique is radviz which puts each feature 
# as a point on a 2D plane, and then simulates having each sample attached to those points through 
# a spring weighted by the relative value for that feature
from pandas.tools.plotting import radviz
radviz(iris_df, "Species")
```
